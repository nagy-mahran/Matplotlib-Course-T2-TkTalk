# First Chart Pipeline

## A. Get your data

### Direct list

```
Duration = ['Jan-20', 'Feb-20', 'Mar-20', 'Apr-20', 'May-20', 'Jun-20', 'Jul-20', 'Aug-20', 'Sep-20', 'Oct-20', 'Nov-20', 'Dec-20', 'Jan-21', 'Feb-21', 'Mar-21']
Monthly = [0,	25,	42,	113,	294,	718,	1497,	2311,	2311,	1497,	718,	294,	113,	42,	21]
Accumulated = [0.00,	0.00,	1.00,	2.00,	5.00,	12.00,	27.00,	50.00,	73.00,	88.00,	95.00,	98.00,	99.00,	100.00,	100.00]
```

```
print("Duration type is ", type(Duration))
print("Duration = ", Duration)
```

### Import from file

```
import pandas as pd

googleSheetId = '1r7chogkaEL6qQBOTLWY1WrGPy6LdvQWhWi_ligNCXAA'
worksheetName = 'Sheet1'
url = 'https://docs.google.com/spreadsheets/d/{0}/gviz/tq?tqx=out:csv&sheet={1}'.format(
    googleSheetId,
    worksheetName
    )
df = pd.read_csv(url)
df
```

### Assign X and Y axes for your charts

```
x = Duration
y_line = Accumulated
y_bar = Monthly
```

## B. Create Figure and Axes

```
from matplotlib import pyplot as plt
plt.ioff()

# Instantiation of Figure and Axes objects
fig, ax = plt.subplots(figsize=(10, 8))

# Define your charts
ax.plot(x, y_line, 'r-', label="Accumulated")
ax2 = ax.twinx()
ax2.bar(x, y_bar, label="Monthly Man-hrs")
```

## C. Spines, Ticks and Grid

```
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

# Configure your chart's spines
ax2.spines['left'].set_color('red')
ax2.spines['right'].set_color('blue')
ax2.spines['bottom'].set_color('green')
ax2.spines['top'].set_color('green')

# Set Ticks parameters
ax.tick_params(axis='x', top=True)
ax.tick_params(axis='y', labelcolor="red")
ax.set_ylim(ymin=0)

ax2.tick_params(axis='y', labelcolor="blue")

# Set minor ticks. Two intervals between majors
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

# Enable chart grid
ax.grid(True, which='major',axis='both', color='#C3C3C3')
ax.grid(True, which='minor',axis='y', color='#EEEEEE')
```

## D. Titles, labels, and legend

```
# Figure Title
fig.suptitle("Project Baseline and Progress")

# Subplot Title
ax.set_title("S-Curve")

# X and Y Axis Labels
ax.set_xlabel("Duration")
ax.set_ylabel("Accumulated Progress", color="red")
ax2.set_ylabel("Monthly Man-hours", color="blue")

# Legend
ax.legend()
ax2.legend()
```

## E. Print your graph on screen

```
plt.show()
```

# Plot directly from Pandas

```
ax = df.plot(x='Date', y='Accumulated%', kind="line", color="red", label="Accumulated Progress",figsize=(10, 8))
df.plot(x='Date', y='Manpower Hrs', kind="bar", label="Monthly Man-hours", secondary_y=True, ax=ax)
plt.title('S-Curve')
plt.xlabel('Duration')
plt.ylabel('Monthly Man-hours')
ax.set_ylabel("Accumulated Progress")

plt.show()
```
